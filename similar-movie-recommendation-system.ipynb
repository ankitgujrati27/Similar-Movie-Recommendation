# Similar Movie Recommendation System (Item-based):

Movie database: http://grouplens.org/datasets/movielens/100k/

```python
import pandas as pd
```

```python
# Importing Movie and Rating Database:

r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('F:\Mine\My Learning\Similar Movie Recommendation\DataSets\ml-100k\u.data',sep = '\t', names=r_cols, usecols=range(3))

m_cols = ['movie_id', 'title']
movies = pd.read_csv('F:Mine\My Learning\Similar Movie Recommendation\DataSets\ml-100k\u.item', sep='|', names=m_cols, usecols=range(2))
```

```python
# Merging the two tables:

ratings = pd.merge(movies,ratings, on = 'movie_id')
```

```python
ratings.head()
```

```python
# Using the pivot_table functions creating a Users/Movie Matrix.

movieRatings = ratings.pivot_table(index='user_id',columns='title',values='rating')
movieRatings.head()
```

```python
# Series of users who rating Star wars:

StarWarsRating = movieRatings['Star Wars (1977)']
StarWarsRating.head()
```

```python
#  Computing the pairwise correlation of Star Wars' vector of user rating with every other movie:

similarMovies = movieRatings.corrwith(StarWarsRating)

# Removing NA values:

similarMovies = similarMovies.dropna()

similarMovies
```

```python
# Sorting the above results:

similarMovies.sort_values(ascending=False)
```

Results are probably getting messed up by movies that have only been viewed by a handful of people who also happened to like Star Wars. Constructing a new DataFrame that counts up how many ratings exist for each movie, and also the average rating.

```python
import numpy as np
```

```python
# Couting number of users rated a particular movie and avg rating:

movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head()
```

```python
# Getting rid of any movies rated by fewer than 100 people, and check the top-rated ones that are left:

popularMovies = movieStats['rating']['size'] >= 100

movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
# Joining this data with our original set of similar movies to Star Wars:

df = movieStats[popularMovies].join(pd.DataFrame(similarMovies, columns=['similarity']))
```

```python
df.head()
```

```python
# Sorting the Data with similarity score:

df.sort_values(['similarity'], ascending=False)[:15]
```
